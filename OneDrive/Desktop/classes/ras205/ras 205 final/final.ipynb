```python
# PROBLEM 1
def surface_area(r):
    pi = 3.1415
    if r <= 0:
        print("Error: Radius must be greater than 0.")
        return -1
    else:
        area = 4 * pi * r ** 2
        return area
        
result = surface_area(5)
print(f"Area with radius 5: {result}")

result = surface_area(10)
print(f"Area with radius 10: {result}")

result = surface_area(15)
print(f"Area with radius 15: {result}")

result = surface_area(20)
print(f"Area with radius 20: {result}")
```

```python
# PROBLEM 2
import numpy as np

def replace(arr):
    if arr.shape[0] != arr.shape[1]:
        print("Array has to be a square")
        return arr
    diagonals = np.diag(arr)
    arr[0] = diagonals
    return arr

original_array = np.random.randint(1, 10, size = (3, 3))
print("Original Array \n", original_array)

modified_array = replace(original_array.copy())
print("\n Modified Array \n", modified_array)
```

```python
# PROBLEM 3
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(-np.pi/2, np.pi/2, 500)

x = (np.cos(theta))**2 * np.sin(theta)
y = (np.sin(theta))**2 * np.cos(theta)

plt.figure(figsize = (8, 8))
plt.plot(x, y, label = r'function')
plt.title('Problem 3 Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')
plt.show()

```

```python
# EXTRA CREDIT
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

c = 0.2
d = 0.5

x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
x, y = np.meshgrid(x, y)

z = (x**2 / c**2) + (y**2/ d**2)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z)

ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
ax.set_title('Extra Credit: Elliptic Paraboloid');
```
